#### main.py

```python
import kivy
from kivy.app import App
from kivy.uix.popup import Popup
from kivy.lang import Builder
from kivy.uix.screenmanager import ScreenManager, Screen
from kivy.uix.widget import Widget
from kivy.uix.floatlayout import FloatLayout


class MainWindow(Screen,Widget):
    def cek(self):
        username = self.ids.username
        username_text = username.text

        password = self.ids.passw
        password_text = str(password.text)
        if username_text == "admin" and password_text == "123":
            self.parent.current = 'second'
        else:
            show_popup()

class SecondWindow(Screen):
    pass

class WindowManager(ScreenManager):
    pass

class layout(FloatLayout):
    pass

def show_popup():
    show = layout()
    popupWindow = Popup(title = "Popup Window", content=show, size_hint=(None,None), size=(400,400))
    popupWindow.open()

kv = Builder.load_file("style.kv")

class MyMainApp(App):
    def build(self):
        return kv

if __name__ == "__main__":
    MyMainApp().run()
```

Didalam script terdapat 2 buah class yang menggunakan class screen yaitu MainWindow dan SecondWindow. #### style.kv

```python
#:import C kivy.utils.get_color_from_hex
WindowManager:
    MainWindow:
    SecondWindow:

<MainWindow>:
    name: "main"
    canvas:
        Color:
            rgba: C('#3f92db')
        Rectangle:
            pos: self.pos
            size: self.size
    FloatLayout:
        Label:
            text: "Login"
            size_hint: 0.8, 0.2
            pos_hint: {"x":0.1, "top":1}
            font_size: (root.width**2 + root.height**2) / 13**4

        Label:
            text:"Email: "
            font_size: (root.width**2 + root.height**2) / 13**4
            pos_hint: {"x":0.06, "top":0.7}
            size_hint: 0.35, 0.1


        TextInput:
            id: username
            font_size: (root.width**2 + root.height**2) / 13**4
            multiline: False
            pos_hint: {"x": 0.45 , "top":0.7}
            size_hint: 0.4, 0.1

        Label:
            text:"Password: "
            font_size: (root.width**2 + root.height**2) / 13**4
            pos_hint: {"x":0.1, "top":0.5}
            size_hint: 0.35, 0.1

        TextInput:
            id: passw
            font_size: (root.width**2 + root.height**2) / 13**4
            multiline: False
            password: True
            pos_hint: {"x": 0.45, "top":0.5}
            size_hint: 0.4, 0.1

        Button:
            pos_hint:{"x":0.2,"y":0.05}
            size_hint: 0.6, 0.2
            font_size: (root.width**2 + root.height**2) / 13**4
            text: "Login"
            on_release:
                root.cek()
                root.manager.transition.direction = "left"


<layout>:
    Label:
        text: "Email atau Password yang anda masukan salah."
        size_hint: 0.6, 0.2
        pos_hint:{"x": 0.2, "top": 1}

    Button:
        text: "Back"
        size_hint: 0.8, 0.2
        pos_hint:{"x": 0.1, "y": 0.1}
        on_release: app.root.current = "main"

<SecondWindow>:
    name: "second"
    Button:
        text: "Succes Login"
        on_release:
            app.root.current = "main"
            root.manager.transition.direction = "right"
```
